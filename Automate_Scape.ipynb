```python
import csv
import pandas as pd



def scrape(dining_hall):
  date = datetime.datetime.utcnow()
  utc_time = calendar.timegm(date.utctimetuple())
  link = "https://5cmenu-cache.jojodmo.com/v1/getMenu/?diningHall=" + str(dining_hall) + "&startTime=" + str(utc_time) + "&language=en"
  response = requests.get(link, verify=False)
  data = response.json()
  return data



result_frary = scrape('frary')
result_frank = scrape('frank')
result_mudd = scrape('Hoch')
result_scripps= scrape('malott')
result_pitzer = scrape('mcconnel')
result_claremont = scrape('collins')



def utc_to_day(utc_timestamp):
    # Convert the UTC timestamp to a datetime object
    utc_datetime = datetime.datetime.utcfromtimestamp(utc_timestamp-28800) 

    # Extract the date from the datetime object and format it as a string
    day_string = utc_datetime.strftime('%Y-%m-%d')

    # Return the date string
    return day_string

curr_date = utc_to_day(utc_time)
```
