```
import os
import pandas as pd
from google.oauth2 import service_account
from google.auth.transport.requests import AuthorizedSession
import google.auth.exceptions
import google.auth.httplib2
import httplib2

# replace with your own credentials file path
credentials_file = os.environ.get("GOOGLE_APPLICATION_CREDENTIALS", "/work/camp-claremont-eef0a71b208c.json")

# create a credentials object from the service account file
creds = service_account.Credentials.from_service_account_file(credentials_file, scopes=["https://www.googleapis.com/auth/drive"])

# create an authorized session object
try:
    session = AuthorizedSession(creds)
except google.auth.exceptions.TransportError as error:
    print(f"An error occurred: {error}")
    session = None

# ID of the file to update
file_id = "1MNW8xJ5l9aj0njmdDuACcFy6Aa7BGQ3g"

# export the file as CSV
export_url = f"https://drive.google.com/uc?id={file_id}&export=download"
resp = session.get(export_url)

# read the downloaded CSV file into a pandas dataframe
df = pd.read_csv(resp.content.decode("utf-8"))

# create a new row to add to the dataframe
df_row = pd.DataFrame([["John", "Doe", "johndoe@gmail.com"]], columns=["First Name", "Last Name", "Email"])

# add the new row to the dataframe
df = df.append(df_row, ignore_index=True)

# convert the modified dataframe back to a CSV file and upload to Google Drive
file_metadata = {"name": "updated_file.csv", "parents": [], "mimeType": "text/csv"} # modify the file name and MIME type if needed
media = pd.DataFrame.to_csv(df, index=False).encode("utf-8")
try:
    resp = session.put(f"https://www.googleapis.com/upload/drive/v3/files/{file_id}?uploadType=media",
                       headers=file_metadata, data=media)
    print(f"File updated: {resp.json()['name']}")
except google.auth.exceptions.TransportError as error:
    print(f"An error occurred: {error}")
    resp = None
```
